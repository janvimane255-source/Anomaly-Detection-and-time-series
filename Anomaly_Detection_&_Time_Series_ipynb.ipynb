Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Ans-


- **Anomaly detection is the process of identifying data points, events, or patterns that significantly differ from the majority of the data, often indicating errors, rare events, or potential issues.**
  
  
  
### What is Anomaly Detection?  
Anomaly detection, also known as outlier detection, is a technique used in **data analysis and machine learning** to identify unusual patterns that do not conform to expected behavior. These anomalies may indicate **fraud, system faults, network intrusions, or rare events**. Detecting anomalies is crucial in areas like finance, cybersecurity, manufacturing, and healthcare.
### Types of Anomaly Detection  
  
**1. Point Anomalies:**  
A point anomaly occurs when a single data instance is significantly different from the rest of the dataset. This type focuses on **individual anomalous points** without much context.  
- Example: A sudden spike in a bank account transaction that is unusually large compared to past transactions.  
  
**2. Contextual (Conditional) Anomalies:**  
In contextual anomaly detection, a data point is considered anomalous based on **its context or environment**. The same data value may be normal in one context but anomalous in another.  
- Example: A temperature reading of 30°C may be normal in summer but anomalous during winter, considering the time of year as context.  
  
**3. Collective Anomalies:**  
These occur when a **collection of related data points collectively behave anomalously**, even if individual points may seem normal. This type is important for detecting patterns that only appear abnormal when viewed together.  
- Example: A sequence of network requests that individually appear normal but in combination indicate a potential cyberattack.  
  
### Summary  
Anomaly detection helps in **identifying deviations from normal patterns** in data, improving decision-making and risk mitigation. The main types are **point anomalies** (individual outliers), **contextual anomalies** (outliers dependent on context), and **collective anomalies** (patterns of anomalies across multiple data points). Understanding these types helps in choosing appropriate **detection algorithms**, such as statistical methods, machine learning models, or neural network approaches, depending on the use case.





Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.


Ans-

-   
  
## 1. Isolation Forest (iForest)  
**Approach:**
- **Tree-based partitioning:** Randomly selects a feature and a split value to recursively partition the dataset.  
- **Isolation principle:** Anomalies are rare and distinct, so they are isolated quickly in fewer splits, resulting in shorter path lengths in the forest of trees.  
- **Anomaly scoring:** Average path length across trees gives an anomaly score; shorter path → higher probability of being an outlier.  
  
**Strengths / Suitable Use Cases:**
- Efficient for **large and high-dimensional datasets**.  
- Detects **global anomalies**.  
- Works well even if the dataset lacks clear clusters or density structures.  
- Common in **fraud detection, intrusion detection, sensor fault analysis**.  
- **Scalable and fast** (O(n)) for real-time applications.  
  
**Limitations:**
- Less effective at detecting subtle **local anomalies** in variable-density regions.  
- Sensitive to the **proportion parameter (contamination)** for expected anomalies.  
  
## 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)  
**Approach:**
- **Density-based clustering:** Groups points into clusters based on high-density regions.  
- **Outlier detection:** Points that lie in low-density areas (noise) are marked as anomalies.  
- Parameters: `eps` (neighborhood radius) and `MinPts` (minimum points to form a dense region) govern clustering.  
  
**Strengths / Suitable Use Cases:**
- Detects **noise and outliers** in datasets with clear **dense and sparse regions**.  
- Handles **arbitrary-shaped clusters**, unlike k-means.  
- Effective for **spatial, time-series, and geospatial data**.  
- Performs well when the data is **pre-processed** or normalized.  
  
**Limitations:**
- Less suitable for **very large datasets** due to computational cost.  
- Sensitive to **parameter selection** (`eps`, `MinPts`).  
- Struggles with varying density: anomalies in dense regions can be missed.  
  
## 3. Local Outlier Factor (LOF)  
**Approach:**
- **Local density estimation:** Compares the density around a data point with that of its neighbors.  
- **Outlier score:** If a point's density is significantly lower than that of its neighbors, it is classified as an anomaly.  
  
**Strengths / Suitable Use Cases:**
- Detects **local outliers** (subtle anomalies relative to neighborhood).  
- Works well in **datasets with varying density**.  
- Preferable for **small to medium datasets**.  
- Useful in **financial fraud detection, medical diagnostics**, and environments where local deviations matter more than global rarity.  
  
**Limitations:**
- Computational complexity **O(n²)**, which reduces scalability for large datasets.  
- Sensitive to **choice of k (number of neighbors)**.  
- Less effective for **global anomalies** that are far from all clusters.  
  
## Comparative Summary  
  
  
| Feature / Algorithm | Isolation Forest | DBSCAN | Local Outlier Factor (LOF) |  
|--------------------|----------------|--------|----------------------------|  
| **Method** | Random tree partitioning | Density-based clustering | Local density comparison |  
| **Anomaly Type** | Global anomalies | Noise in low-density regions | Local anomalies relative to neighbors |  
| **Dataset Size Suitability** | Large to very large | Small to medium, pre-processed | Small to medium |  
| **Dimensionality Handling** | High-dimensional data | Moderate dimensions, spatially meaningful | Moderate dimensions |  
| **Computational Complexity** | O(n) | Usually O(n log n)–O(n²) depending on index | O(n²) |  
| **Cluster Shape Handling** | Independent of clusters | Arbitrary shapes | Depends on neighborhood definition |  
| **Use Cases** | Fraud detection, sensor faults, real-time monitoring | Spatial clustering, geolocation data, pre-clustered data | Local anomaly detection, subtle fraud, quality control |  
  
  
### Practical Insights  
- **Isolation Forest** is preferred for **large, high-dimensional streaming data**, where **real-time anomaly detection** is critical.  
- **DBSCAN** is useful when **cluster structure and density contrasts** define normal vs. anomalous points.  
- **LOF** excels in detecting **subtle or local irregularities**, especially in **heterogeneous density regions**, but may not scale well.  
  


Question 3: What are the key components of a Time Series? Explain each with one
example.

Ans-

- **The main components of a time series are trend, seasonality, cyclical patterns, and irregular variations, each representing distinct patterns in the data that help in analysis and forecasting** [^1^] [^2^] [^5^].
  
  
  
### 1. Trend  
**Trend** represents the long-term movement or direction of the data over time. It indicates whether the values are generally increasing, decreasing, or remaining stable across a period. For example, the annual revenue of a company showing consistent growth over five years indicates an upward trend [^9^].
### 2. Seasonality  
**Seasonality** refers to regular, repeating patterns at fixed intervals, such as daily, weekly, monthly, or yearly cycles. These patterns are typically influenced by external factors like holidays, climate, or social customs. An example is sales of ice cream that peak every summer or retail sales surging during holiday seasons [^9^].
### 3. Cyclical Patterns  
**Cyclical patterns** are fluctuations that occur over longer, often unpredictable periods, differing from seasonality as they do not follow a fixed calendar schedule. These are often tied to broader economic or business cycles, such as periods of economic recession and expansion lasting multiple years [^7^] [^9^].
### 4. Irregular Variations (Noise)  
**Irregular variations**, also called residuals or noise, are random, unpredictable fluctuations in the data caused by unforeseen events, measurement errors, or other unique factors. For instance, sudden natural disasters affecting production data represent irregular variations. These are not part of the trend, seasonality, or cycles and are often isolated when building forecasting models [^9^].
By decomposing a time series into these four components, analysts can **better understand the underlying patterns, make more accurate forecasts, and identify anomalies**. Various methods like additive or multiplicative decomposition are used to separate these components effectively [^9^].




Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

Ans-    
  
In time series analysis, a time series is said to be **stationary** if its **statistical properties do not change over time**. This implies that aspects such as mean, variance, and autocorrelation structure remain constant and do not depend on the specific time at which the data is observed.  
Formally, there are different levels:
1. **Strict Stationarity**: The joint probability distribution of any set of observations is invariant under time shifts. That is, for all \\(t_1, t_2, \\dots, t_k\\) and any lag \\(h\\):  
  
   \\[
F_{X_{t_1}, \\dots, X_{t_k}}(x_1, \\dots, x_k) = F_{X_{t_1+h}, \\dots, X_{t_k+h}}(x_1, \\dots, x_k)
\\]
2. **Weak (Second-Order) Stationarity**: Common in practice, requiring only the first two moments to be time-invariant:  
   - Constant mean: \\(\\mathbb{E}[X_t] = \\mu\\)  
   - Constant variance: \\(\\mathrm{Var}(X_t) = \\sigma^2 < \\infty\\)  
   - Covariance depends only on the lag: \\(\\mathrm{Cov}(X_t, X_{t+h}) = \\gamma(h)\\)  
  
3. **Trend Stationarity**: The series may have a deterministic trend. After removing the trend, the series becomes stationary.  
  
4. **Difference Stationarity**: The series becomes stationary after one or more differences.  
  
# Importance of Stationarity  
  
- **Predictability**: Stationary series are easier to model, as their future behavior depends only on statistical properties that remain consistent.  
- **Model assumptions**: Many time series models (ARIMA, ARMA) assume stationarity; failing this assumption can lead to spurious results.  
- **Reliable inference**: Non-stationary series may give misleading statistical measures (mean, variance, correlations).  
  
# Testing for Stationarity  
  
## 1. Visual Methods  
- **Time Plot**: Observe constancy of mean and variance over time.  
- **Rolling Statistics**: Compute rolling mean and variance; large changes suggest non-stationarity.  
- **Autocorrelation Function (ACF)**: Non-decaying autocorrelations indicate non-stationarity.  
  
## 2. Summary Statistics  
- Compare mean and variance across different segments of the series. Significant differences suggest non-stationarity.  
  
## 3. Statistical Tests  
- **Augmented Dickey-Fuller (ADF) Test**:  
  - Null: The series has a unit root (non-stationary)  
  - Alternative: Series is stationary  
  - Rejecting null (p-value \\(\\le 0.05\\)) implies stationarity.  
  
- **KPSS Test**:  
  - Null: Series is stationary  
  - Alternative: Series has a unit root  
  - Rejecting null (p-value \\(\\le 0.05\\)) indicates non-stationarity.  
  
- **Phillips-Perron (PP) Test**: Similar to ADF, adjusted for autocorrelation and heteroskedasticity.  
  
# Transforming Non-Stationary Time Series  
  
If a series is non-stationary, you can apply transformations:
1. **Differencing**:  
   - First difference: \\(\\Delta X_t = X_t - X_{t-1}\\)  
   - Seasonal difference: \\(\\Delta_s X_t = X_t - X_{t-s}\\)  
  
2. **Detrending**:  
   - Remove linear or nonlinear deterministic trends, keeping residuals.  
  
3. **Power or Log Transformations**:  
   - Stabilize variance (log, square root, Box-Cox).  
   - Box-Cox transformation:  
  
     \\[
X_t^{(\\lambda)} =
\\begin{cases}
\\frac{X_t^\\lambda - 1}{\\lambda}, & \\lambda
eq 0 \\\\
\\ln X_t, & \\lambda = 0
\\end{cases}
\\]
4. **Deseasonalization**:  
   - Remove seasonal patterns using decomposition techniques.  
  
After applying transformations, re-test for stationarity to ensure the series is adequately stationary for modeling.
# Conclusion  
  
Stationarity is a cornerstone in time series analysis. Testing (ADF, KPSS, PP) and transforming (differencing, detrending, log, Box-Cox) non-stationary series is essential before modeling, especially for ARIMA and similar models, to achieve accurate predictions and reliable inference.




Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.


Ans-   
  
Time series analysis often involves modeling patterns in sequential data to make forecasts. Several autoregressive and moving average based models form the backbone of classical statistical approaches. Here is a detailed comparison and differentiation of **AR, MA, ARIMA, SARIMA, and SARIMAX models**:
### 1. AR (Autoregressive) Model  
- **Definition**: The AR model predicts a variable using a linear combination of its past values.  
- **Mathematical Formulation**:    
  
  \\[
X_t = c + \\sum_{i=1}^{p} \\phi_i X_{t-i} + \\epsilon_t
\\]  
where:  
\\(X_t\\) = current value,  
\\(c\\) = constant term,  
\\(\\phi_i\\) = autoregressive coefficients,  
\\(p\\) = order of the model (number of lagged terms),  
\\(\\epsilon_t\\) = white noise error term.
- **Key Feature**: Captures **momentum or persistence** in the series; relies solely on past values.  
- **Use Case**: Stock prices, economic indicators with self-correlated patterns.  
  
### 2. MA (Moving Average) Model  
- **Definition**: The MA model expresses the variable as a linear combination of past **error terms** rather than past values.  
- **Mathematical Formulation**:    
  
  \\[
X_t = \\mu + \\epsilon_t + \\sum_{i=1}^{q} \\theta_i \\epsilon_{t-i}
\\]  
where:  
\\(\\epsilon_t, \\epsilon_{t-i}\\) = white noise errors,  
\\(\\theta_i\\) = MA coefficients,  
\\(q\\) = order of the MA process.
- **Key Feature**: Smooths out short-term shocks; captures **noise correlation** rather than autocorrelation.  
- **Use Case**: Time series with irregular shocks or random fluctuations.  
  
### 3. ARIMA (Autoregressive Integrated Moving Average) Model  
- **Definition**: ARIMA generalizes AR and MA by including **differencing** to make a non-stationary series stationary.  
- **Components**:    
  - **AR(p)**: number of autoregressive terms    
  - **I(d)**: number of differences needed to achieve stationarity    
  - **MA(q)**: number of moving average terms  
- **Mathematical Formulation**:    
  
  If \\(Y_t = \\Delta^d X_t\\) is the differenced series:  
\\[
Y_t = c + \\sum_{i=1}^{p} \\phi_i Y_{t-i} + \\sum_{j=1}^{q} \\theta_j \\epsilon_{t-j} + \\epsilon_t
\\]  
- **Key Feature**: Handles **trends** and autocorrelated errors; essential for non-stationary time series.  
- **Use Case**: Macroeconomic data, demand forecasting.  
  
### 4. SARIMA (Seasonal ARIMA) Model  
- **Definition**: SARIMA extends ARIMA to **incorporate seasonality** using seasonal AR and MA terms along with seasonal differencing.  
- **Components**: \\(ARIMA(p,d,q)(P,D,Q)_s\\)    
  - \\(P, D, Q\\) = seasonal AR, differencing, and MA terms    
  - \\(s\\) = seasonal periodicity (e.g., 12 for monthly data with yearly seasonality)  
- **Mathematical Formulation**:    
  
  Combines non-seasonal ARIMA with seasonal terms:
\\[
\\Phi_P(B^s) \\phi_p(B) \\Delta^d \\Delta_s^D X_t = \\Theta_Q(B^s) \\theta_q(B) \\epsilon_t
\\]  
where \\(B\\) is the backshift operator, and \\(\\Delta_s\\) represents seasonal differencing.
- **Key Feature**: Models repeated seasonal effects in addition to trend and autocorrelation.  
- **Use Case**: Monthly sales with yearly patterns, seasonal climate data.  
  
### 5. SARIMAX (SARIMA with eXogenous variables)  
- **Definition**: SARIMAX further extends SARIMA by including **exogenous (external) regressors** to account for additional influences on the dependent variable.  
- **Mathematical Formulation**:    
  
  \\[
X_t = \\text{SARIMA terms} + \\mathbf{\\beta}^T \\mathbf{Z}_t + \\epsilon_t
\\]  
where \\(\\mathbf{Z}_t\\) are exogenous inputs (e.g., marketing spend, holidays).
- **Key Feature**: Combines autoregressive, moving average, seasonal modeling **and influence of external factors**.  
- **Use Case**: Forecasting demand considering advertising campaigns, weather, or other external covariates.  
  
### Comparison Table  
  
  
| Model     | Uses Past Values? | Uses Past Errors? | Handles Non-Stationarity? | Handles Seasonality? | Handles Exogenous Variables? |  
|-----------|-----------------|-----------------|--------------------------|-------------------|-----------------------------|  
| AR        | Yes             | No              | No                       | No                | No                          |  
| MA        | No              | Yes             | No                       | No                | No                          |  
| ARIMA     | Yes             | Yes             | Yes (via differencing)   | No                | No                          |  
| SARIMA    | Yes             | Yes             | Yes                      | Yes               | No                          |  
| SARIMAX   | Yes             | Yes             | Yes                      | Yes               | Yes                         |  
  
  
### Summary  
- **AR** → relies only on past values    
- **MA** → relies only on past errors    
- **ARIMA** → combines AR + MA + differencing for stationary transformation    
- **SARIMA** → adds seasonal components to ARIMA    
- **SARIMAX** → adds external regressors to SARIMA, suitable for time series influenced by covariates    
  
This hierarchy represents increasing **flexibility and complexity**, from simple autocorrelation modeling to seasonal and multi-factor predictive modeling.


Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. Generate a synthetic numerical dataset
# Generate 'normal' data points
np.random.seed(42)
X_normal = np.random.randn(300, 2) * 2 + 5  # Centered around (5,5)

# Generate some 'outlier' data points
X_outliers = np.random.rand(30, 2) * 10 + 15  # Far from normal data, centered around (20,20)

# Add some subtle outliers within the normal distribution
X_subtle_outliers = np.random.randn(20, 2) * 5 + np.array([-10, -10]) # Far from normal data, centered around (-10,-10)

X = np.vstack([X_normal, X_outliers, X_subtle_outliers])

df_synthetic = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])

print("Synthetic Dataset Head:")
print(df_synthetic.head())
print(f"\nSynthetic Dataset Shape: {df_synthetic.shape}")
```

```python
# 2. Apply Isolation Forest
# The contamination parameter is the proportion of outliers in the dataset.
# We can estimate it or use a default value if unknown.
# Here, we know we generated 50 outliers out of 350 total points (approx 14%)
model = IsolationForest(n_estimators=100, max_samples='auto', contamination=0.14, random_state=42)
model.fit(df_synthetic[['Feature_1', 'Feature_2']])

# Get the raw anomaly scores (lower score indicates higher anomaly likelihood)
df_synthetic['scores'] = model.decision_function(df_synthetic[['Feature_1', 'Feature_2']])

# Predict the anomaly scores (-1 for outliers, 1 for inliers)
df_synthetic['anomaly'] = model.predict(df_synthetic[['Feature_1', 'Feature_2']])

# Separate anomalies and normal points for visualization
anomalies = df_synthetic[df_synthetic['anomaly'] == -1]
normal_points = df_synthetic[df_synthetic['anomaly'] == 1]

print("\nNumber of detected anomalies:", len(anomalies))
print("Detected Anomalies Head:")
print(anomalies.head())
```

```python
from sklearn.neighbors import LocalOutlierFactor

# 1. Apply Local Outlier Factor (LOF)
# n_neighbors: number of neighbors to consider for the local density calculation
# contamination: the proportion of outliers in the dataset. Can be estimated or set.
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.15) # Increased contamination slightly from IF

# Fit the model and predict (LOF returns -1 for outliers and 1 for inliers)
# Note: LOF.fit_predict returns the anomaly labels directly
df_synthetic['lof_anomaly'] = lof.fit_predict(df_synthetic[['Feature_1', 'Feature_2']])

# Get the raw anomaly scores (negative_outlier_factor_)
# The lower the lof_score, the more likely the point is an outlier.
# LOF convention: scores closer to -1 are normal, significantly lower than -1 are outliers.
df_synthetic['lof_scores'] = lof.negative_outlier_factor_

# Separate anomalies and normal points for visualization based on LOF results
lof_anomalies = df_synthetic[df_synthetic['lof_anomaly'] == -1]
lof_normal_points = df_synthetic[df_synthetic['lof_anomaly'] == 1]

print("\nNumber of detected LOF anomalies:", len(lof_anomalies))
print("Detected LOF Anomalies Head:")
print(lof_anomalies.head())
```

```python
# 2. Visualize the LOF anomalies on a 2D scatter plot
plt.figure(figsize=(12, 8))

# Plot normal points in blue
plt.scatter(lof_normal_points['Feature_1'], lof_normal_points['Feature_2'],
            c='blue', s=20, label='Normal Points', alpha=0.7)

# Plot anomalies in red
plt.scatter(lof_anomalies['Feature_1'], lof_anomalies['Feature_2'],
            c='red', s=50, marker='o', label='Anomalies', edgecolors='k', linewidths=1)

# Optionally, add the decision boundary (if needed, but harder to visualize for LOF without a grid)

plt.title('Anomaly Detection using Local Outlier Factor (LOF)', fontsize=16)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# The 'data' DataFrame from previous cells contains the AirPassengers dataset.
# Ensure the index is datetime and the column is named 'Passengers'

# 1. Determine SARIMA orders (often found through ACF/PACF analysis, or common knowledge for AirPassengers)
# For AirPassengers, a common order is (1,1,1) for non-seasonal and (1,1,1,12) for seasonal
# Non-seasonal: (p,d,q) = (1,1,1)
# Seasonal: (P,D,Q,s) = (1,1,1,12)

order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12)

# 2. Train the SARIMA model
model = SARIMAX(data['Passengers'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit(disp=False)

print(model_fit.summary())
```

```python
# 3. Forecast the next 12 months
# Get the last date in the training data
last_date = data.index[-1]

# Generate future dates for the forecast (next 12 months)
forecast_index = pd.date_range(start=last_date, periods=13, freq='MS')[1:] # Exclude the start date itself

# Make predictions
forecast = model_fit.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# 4. Visualize the results
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Passengers'], label='Original Series', color='blue')

# Plot the fitted values
plt.plot(model_fit.fittedvalues.index, model_fit.fittedvalues, label='SARIMA Fitted Values', color='green', linestyle='--')

# Plot the forecast
plt.plot(forecast_index, forecast.predicted_mean, label='Forecast (Next 12 Months)', color='red')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3, label='95% Confidence Interval')

plt.title('Monthly Air Passengers: SARIMA Forecast', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

```python
# Python

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Generate a sample dataset with 2D numerical features
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=1.0, random_state=42)

# Add some outliers manually
np.random.seed(42)
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, outliers])

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # contamination is the expected outlier fraction
y_pred = lof.fit_predict(X)  # -1 for outlier, 1 for inlier
lof_scores = -lof.negative_outlier_factor_  # higher score = more anomalous

# Separate inliers and outliers for visualization
inliers = X[y_pred == 1]
outliers = X[y_pred == -1]

# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], color='blue', label='Inliers')
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', marker='x', label='Outliers')
plt.title("Local Outlier Factor (LOF) Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Print LOF scores for reference
for i, score in enumerate(lof_scores):
    print(f"Point {i}: LOF Score = {score:.2f}, {'Outlier' if y_pred[i]==-1 else 'Inlier'}")
```

## Understanding Power Grid Data    
  
Power grids generate vast amounts of data from different sources:  
- **Smart meters** provide real-time consumption and voltage data.    
- **Supervisory Control and Data Acquisition (SCADA) systems** monitor grid operations, such as power flows, transformer status, and line loads.    
- **Sensors and IoT devices** gather environmental parameters, like temperature, humidity, and weather, which affect grid performance.    
  
Data may be structured, semi-structured, or streaming, requiring preprocessing to handle missing or noisy values, outlier detection, and normalization.
## Key Responsibilities    
  
1. **Data Integration and Cleaning:**    
   - Transform raw data from multiple sources into a unified schema.    
   - Handle missing data, outliers, and potential sensor errors using statistical or machine learning techniques.  
  
2. **Descriptive Analytics:**    
   - Summarize historical energy consumption, power flows, and disruptions.    
   - Generate dashboards and visualizations to monitor performance metrics, such as line load capacity and frequency deviations.  
  
3. **Predictive Modeling:**    
   - Forecast electricity demand using regression models or time series techniques like ARIMA, LSTM, or Prophet.    
   - Predict equipment failures or outages with classification models, such as Random Forests or Gradient Boosting, based on sensor readings, environmental factors, and historical failures.  
  
4. **Anomaly Detection:**    
   - Identify unusual behavior that could indicate faults, line overloads, or cyber-attacks. Techniques include statistical thresholds, clustering methods, or unsupervised learning (e.g., autoencoders).  
  
5. **Optimization and Decision Support:**    
   - Help optimize grid operations by recommending load balancing, energy storage utilization, or maintenance scheduling.    
   - Provide insights to reduce energy losses and improve grid reliability.  
  
## Tools and Techniques    
  
- **Programming:** Python or R for data processing, analysis, and machine learning.    
- **Big Data Frameworks:** Apache Spark or Hadoop for managing high-volume streaming data.

- **Machine Learning Libraries:** Scikit-learn, TensorFlow, PyTorch for predictive and unsupervised models.
